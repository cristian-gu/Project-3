```python
# def average(val):
#     temp_list = []
#     temp_list.append(val)
#     counter = len(temp_list)
#     average_val = average_action(counter_temp)
#     return (counter, temp_list)

def average(temp_list):
    average_val = round(np.average(temp_list), 2)
    return (average_val)
```

```python
def ave_op(list_num):
    temp_list = []
    avg_list = []

    for i in range(1, len(list_num)):
        if list_num[i][1] == list_num[i-1][1]:
            temp_list.append(list_num[i-1][0])
        else:
            temp_list.append(list10[i-1][0])
            holder = average(temp_list)
            avg_list.append(holder)
            temp_list = []
    holder=average(temp_list)
    avg_list.append(holder)
    return (avg_list)

#         print(temp_list)
```

```python
date_column = []
date_int = 2010
while date_int <= 2022:
    date_column.append(date_int)
    date_int +=1
    
# print(date_column)

avg_years_list = {}

for i in range(len(years_list)):
    avg_years_list[date_column[i]] = ave_op(years_list[i])
    
print(avg_years_list)
```

```python
## Combine the new dataset
avg_years_df = pd.DataFrame(avg_years_list)
avg_years_df = avg_years_df.reset_index(level=0)
```
